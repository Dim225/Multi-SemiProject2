```python
test_list = []
file_list = os.listdir('음성인식/train')
for i in range(1000):
    text_path = '음성인식/train/' + file_list[i]
    sig, sr =  librosa.load(text_path , sr=16000)

    non_silence_indices = librosa.effects.split(sig, top_db=30)
    signal = np.concatenate([sig[start:end] for start, end in non_silence_indices])

    test_list.append(signal)
```
